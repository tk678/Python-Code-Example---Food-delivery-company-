```python
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_4.8.pkl'))
```

```python
#grouping by the loyalty flag and seeing what the mean of each group is by price 
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
#creating a group on user Id and their mean spend per order
user_avg_price = ords_prods_merge.groupby('user_id')['prices'].mean()
```

```python
#merge that back into the data frame on user id to show their mean spend 
ords_prods_merge = ords_prods_merge.merge(user_avg_price.rename('avg_price'), on='user_id')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head(5)
```

```python
#checking how many there is of high and low 
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Calculating median days since last order using user id
user_median_days = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()
```

```python
# adding that to dataframe
ords_prods_merge = ords_prods_merge.merge(user_median_days.rename('median_days'), on='user_id')
```

```python
#chacking the new column is there 
ords_prods_merge.head(5)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[
    (ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 
    'order_frequency_flag'
] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(5)
```

```python
#checking how many there is of 
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
