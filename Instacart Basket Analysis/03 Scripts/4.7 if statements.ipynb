```python
df = df_ords_prods_merge[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# for the busiest hours of the day i am going to splt the groups up by top 4 buttom 4 and the rest 
result_hours = []

for value in df_ords_prods_merge["hour_of_the_day_for_order"]:
  if value in [10, 11, 14, 15]:
    result_hours.append("Most orders")
  elif value in [4, 3, 2, 5]:
    result_hours.append("Fewest orders")
  else:
    result_hours.append("Average orders")
```
